## 1、Diabetes Prediction 糖尿病预测

```python
import numpy as np
import pandas as pd
import torch
# 1. 加载数据集

# numpy 读取数据

# 我们的数据集有列名，所以要跳过第一行
xy = np.loadtxt('./data/diabetes.csv', delimiter=',', skiprows=1,dtype=np.float32)
x_data = torch.from_numpy(xy[:,:-1])
# torch.from_numpy 会根据ndarray创建Tensor
y_data = torch.from_numpy(xy[:,-1])
print(x_data.shape)
print(y_data.shape)
```

```python
# pandas 读取数据

xy = pd.read_csv('./data/diabetes.csv',dtype=np.float32)

# <class 'numpy.ndarray'>
print(type(xy.values))
x_data = torch.from_numpy(xy.values[:,:-1])
# [-1]和-1的区别 
# -1维度会改变
# y_data = torch.from_numpy(xy.values[:,-1])
# [-1]保持维度不变
y_data = torch.from_numpy(xy.values[:,[-1]])
print(x_data.shape)
print(y_data.shape)

# print(xy.values[1,1].dtype)
```

```python
# 2、定义模型

class Model(torch.nn.Module):
    def __init__(self):
        super().__init__()
        self.linear1 = torch.nn.Linear(8,6)
        self.linear2 = torch.nn.Linear(6,4)
        self.linear3 = torch.nn.Linear(4,1)
        # 定义sigmoid函数

        self.sigmoid = torch.nn.Sigmoid()
        # 其他的激活函数
        self.relu = torch.nn.ReLU()
        
    def forward(self,x):
        x = self.sigmoid(self.linear1(x))
        x = self.sigmoid(self.linear2(x))
        y_pred = self.sigmoid(self.linear3(x))
        return y_pred

model = Model()

# 3、损失函数和优化器
criterion = torch.nn.BCELoss()
optimizer = torch.optim.SGD(model.parameters(),lr=0.1)

# 4、训练循环
for epoch in range(1000):
    # 前向传播
    y_pred = model(x_data)
    loss = criterion(y_pred,y_data)

    # 梯度清零
    optimizer.zero_grad()
    # 反向传播
    loss.backward()

    # 更新参数
    optimizer.step()

    if epoch % 100 == 0:
        print(f'epoch: {epoch}, loss: {loss.item():.4f}')


```
