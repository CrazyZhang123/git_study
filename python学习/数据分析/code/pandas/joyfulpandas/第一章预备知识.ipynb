### 2ã€åŒ¿åå‡½æ•°ä¸mapæ–¹æ³•

æœ‰ä¸€äº›å‡½æ•°çš„å®šä¹‰å…·æœ‰æ¸…æ™°ç®€å•çš„æ˜ å°„å…³ç³»ï¼Œä¾‹å¦‚ä¸Šé¢çš„ my_func å‡½æ•°ï¼Œè¿™æ—¶å€™å¯ä»¥ç”¨åŒ¿åå‡½æ•°çš„æ–¹æ³•ç®€æ´åœ°è¡¨ç¤ºã€‚

enumerate æ˜¯ä¸€ç§ç‰¹æ®Šçš„æ‰“åŒ…ï¼Œå®ƒå¯ä»¥åœ¨è¿­ä»£æ—¶ç»‘å®šè¿­ä»£å…ƒç´ çš„éå†åºå·ã€‚

æ—¢ç„¶æœ‰äº†å‹ç¼©å‡½æ•°ï¼Œé‚£ä¹ˆ Python ä¹Ÿæä¾›äº† * æ“ä½œç¬¦å’Œ zip è”åˆä½¿ç”¨æ¥è¿›è¡Œè§£å‹æ“ä½œã€‚

````python
my_func = lambda x: 2*x
my_func(2)
# è¾“å‡ºï¼š4

# å®é™…ä½¿ç”¨ä¸éœ€è¦åå­—
[(lambda x: 2*x)(i) for i in range(5)]
# è¾“å‡ºï¼š [0, 2, 4, 6, 8]

# å¯¹äºä¸Šè¿°çš„è¿™ç§åˆ—è¡¨æ¨å¯¼å¼çš„åŒ¿åå‡½æ•°æ˜ å°„ï¼Œ
#  Python ä¸­æä¾›äº† map å‡½æ•°æ¥å®Œæˆï¼Œå®ƒè¿”å›çš„æ˜¯ä¸€ä¸ª map å¯¹è±¡ï¼Œ
# éœ€è¦é€šè¿‡ list è½¬ä¸ºåˆ—è¡¨ï¼š
list(map(lambda x: 2*x,range(5)))
# è¾“å‡ºï¼š[0, 2, 4, 6, 8]

# å¤šä¸ªå‚æ•°çš„æƒ…å†µ
list(map(lambda x, y: str(x)+'_'+y, range(5),list('abcdw')))
# è¾“å‡ºï¼š['0_a', '1_b', '2_c', '3_d', '4_w']
````

### 3ã€zipå‡½æ•°ä¸enumerateæ–¹æ³•

zipå‡½æ•°èƒ½å¤ŸæŠŠå¤šä¸ªå¯è¿­ä»£å¯¹è±¡æ‰“åŒ…æˆä¸€ä¸ªå…ƒç»„æ„æˆçš„å¯è¿­ä»£å¯¹è±¡ï¼Œå®ƒè¿”å›äº†ä¸€ä¸ª zip å¯¹è±¡ï¼Œé€šè¿‡ tuple, list å¯ä»¥å¾—åˆ°ç›¸åº”çš„æ‰“åŒ…ç»“æœï¼š

````python
L1, L2, L3 = list('abc'), list('def'), list('ghi')
list(zip(L1,L2,L3))
# è¾“å‡º [('a', 'd', 'g'), ('b', 'e', 'h'), ('c', 'f', 'i')]

tuple(zip(L1,L2,L3))
# è¾“å‡ºï¼š(('a', 'd', 'g'), ('b', 'e', 'h'), ('c', 'f', 'i'))

# åœ¨å¾ªç¯ä¸­ä½¿ç”¨zipå‡½æ•°
for i,j,k in zip(L1,L2,L3):
    print(i,j,k)
# è¾“å‡º 
# a d g
# b e h
# c f i
````

````python
# enumerate æ˜¯ä¸€ç§ç‰¹æ®Šçš„æ‰“åŒ…ï¼Œå®ƒå¯ä»¥åœ¨è¿­ä»£æ—¶ç»‘å®šè¿­ä»£å…ƒç´ çš„éå†åºå·ï¼š
L = list('abcd')

# (1)enumerate å¯ä»¥æŒ‡å®šèµ·å§‹åºå· start=1
for index,value in enumerate(L,1):
    print(index,value)
# è¾“å‡º
# 0 a
# 1 b
# 2 c
# 3 d

# (2)ä½¿ç”¨zipå®ç°
for index,value in zip(range(len(L)),L):
    print(index,value)

# éœ€è¦å¯¹ä¸¤ä¸ªåˆ—è¡¨å»ºç«‹å­—å…¸æ˜ å°„æ—¶ï¼Œä½¿ç”¨zip
dict(zip(L1,L2))
# è¾“å‡ºï¼š{'a': 'd', 'b': 'e', 'c': 'f'}
````

````python
# æ—¢ç„¶æœ‰äº†å‹ç¼©å‡½æ•°ï¼Œé‚£ä¹ˆ Python ä¹Ÿæä¾›äº† * æ“ä½œç¬¦å’Œ zip è”åˆä½¿ç”¨æ¥è¿›è¡Œè§£å‹æ“ä½œï¼š
zipped = list(zip(L1,L2,L3))
# è¾“å‡º ï¼š [('a', 'd', 'g'), ('b', 'e', 'h'), ('c', 'f', 'i')]

list(zip(*zipped))
# å…ˆæŠŠæ¯ä¸€ä¸ªå…ƒç´ æå–å‡ºæ¥ï¼Œæ‰“æ•£ï¼Œç„¶åä½¿ç”¨zipæ‹‰é“¾è¿›è¡Œè¿æ¥ï¼Œæœ€åå¾—åˆ°3ä¸ªåˆ—è¡¨
# è¾“å‡ºï¼š [('a', 'b', 'c'), ('d', 'e', 'f'), ('g', 'h', 'i')]
````

ã€cã€‘ç»´åº¦å˜æ¢ï¼š reshape

reshape èƒ½å¤Ÿå¸®åŠ©ç”¨æˆ·æŠŠåŸæ•°ç»„æŒ‰ç…§æ–°çš„ç»´åº¦é‡æ–°æ’åˆ—ã€‚åœ¨ä½¿ç”¨æ—¶æœ‰ä¸¤ç§æ¨¡å¼ï¼Œåˆ†åˆ«ä¸º*Cæ¨¡å¼(é»˜è®¤)*å’Œ F æ¨¡å¼ï¼Œåˆ†åˆ«ä»¥é€è¡Œå’Œé€åˆ—çš„é¡ºåºè¿›è¡Œå¡«å……è¯»å–ã€‚

````python
target = np.arange(8).reshape(2,4)
# array([[0, 1, 2, 3],
    #    [4, 5, 6, 7]])

# æŒ‰ç…§è¡Œè¯»å–å’Œå¡«å……ï¼ˆé»˜è®¤ï¼‰
target.reshape((4,2),order='C')
# array([[0, 1],
#        [2, 3],
#        [4, 5],
#        [6, 7]])

# æŒ‰ç…§åˆ—è¯»å–å’Œå¡«å……
target.reshape((4,2),order='F')
# array([[0, 2],
    #    [4, 6],
    #    [1, 3],
    #    [5, 7]])

# ç‰¹åˆ«åœ°ï¼Œç”±äºè¢«è°ƒç”¨æ•°ç»„çš„å¤§å°æ˜¯ç¡®å®šçš„ï¼Œ reshape å…è®¸æœ‰ä¸€ä¸ªç»´åº¦å­˜åœ¨ç©ºç¼ºï¼Œæ­¤æ—¶åªéœ€å¡«å……-1å³å¯ï¼š
target.reshape((-1,2))
# array([[0, 1],
    #    [2, 3],
    #    [4, 5],
    #    [6, 7]])

# è½¬æ¢ä¸ºä¸€ç»´æ•°ç»„
target.reshape(-1)
# array([0, 1, 2, 3, 4, 5, 6, 7])
````

ã€bã€‘ nonzero, argmax, argmin

è¿™ä¸‰ä¸ªå‡½æ•°è¿”å›çš„éƒ½æ˜¯ç´¢å¼•ï¼Œ nonzero è¿”å›éé›¶æ•°çš„ç´¢å¼•ï¼Œ argmax, argmin åˆ†åˆ«è¿”å›æœ€å¤§å’Œæœ€å°æ•°çš„ç´¢å¼•ï¼š

ã€dã€‘ cumprod, cumsum, diff

cumprod, cumsum åˆ†åˆ«è¡¨ç¤ºç´¯ä¹˜å’Œç´¯åŠ å‡½æ•°ï¼Œè¿”å›åŒé•¿åº¦çš„æ•°ç»„ï¼Œ diff è¡¨ç¤ºå’Œå‰ä¸€ä¸ªå…ƒç´ åšå·®ï¼Œç”±äºç¬¬ä¸€ä¸ªå…ƒç´ ä¸ºç¼ºå¤±å€¼ï¼Œå› æ­¤åœ¨é»˜è®¤å‚æ•°æƒ…å†µä¸‹ï¼Œè¿”å›é•¿åº¦æ˜¯åŸæ•°ç»„å‡1

````python
a = np.array([1,2,3])

# ç´¯ä¹˜
a.cumprod()
# array([1, 2, 6])

# ç´¯åŠ 
a.cumsum()
# array([1, 3, 6])

# å·®
np.diff(a)
````

ã€eã€‘ ç»Ÿè®¡å‡½æ•°

å¸¸ç”¨çš„ç»Ÿè®¡å‡½æ•°åŒ…æ‹¬ max, min, mean, median, std, var, sum, quantile ï¼Œå…¶ä¸­åˆ†ä½æ•°è®¡ç®—æ˜¯å…¨å±€æ–¹æ³•ï¼Œå› æ­¤**ä¸èƒ½é€šè¿‡ array.quantile çš„æ–¹æ³•è°ƒç”¨**ï¼š
