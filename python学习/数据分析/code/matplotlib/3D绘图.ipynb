
mpl_toolkits 是 Matplotlib 的绘图工具包。

## 1、3D线图
下面编写第一个三维绘图程序。

首先创建一个三维绘图区域， plt.axes() 函数提供了一个参数projection，将其参数值设置为 "3d"。如下所示：

```python
# 导入三维工具包
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
```

有了3D绘图区域，接下来构建3d图像，如下所示：

```python
fig = plt.figure()
# 创建3d绘图区域
ax = plt.axes(projection='3d')
# 从三个维度构建
z = np.linspace(0,1,100)
x = z * np.sin(20 * z)
y = z * np.cos(20 * z)
```

最后调用plot3D()方法绘制3d图像，代码如下：

```python
# 调用 ax.plot3D创建三维线图
ax.plot3D(x,y,z,'gray')
ax.set_title('3D line plot')
plt.show()
```

完整代码

```python
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
fig = plt.figure()
#创建3d绘图区域
ax = plt.axes(projection='3d')
#从三个维度构建
z = np.linspace(0, 1, 100)
x = z * np.sin(20 * z)
y = z * np.cos(20 * z)
#调用 ax.plot3D创建三维线图
ax.plot3D(x, y, z, 'gray')
ax.set_title('3D line plot')
plt.show()
```

上述代码中的ax.plot3D()函数可以绘制各种三维图像，这些三维图像都要根据(x,y,z)三元组来创建。

## 2、3D散点图

通过ax.scatter3D()函数可以绘制3D散点图，示例代码如下：

```python
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt

fig = plt.figure()
# 创建3d绘图区域
ax = plt.axes(projection='3d')
# 构建xyz
z = np.linspace(0,1,100)
x = z * np.sin(20*z)
y = z * np.cos(20*z)
c = x + y
ax.scatter3D(x,y,z,c=c)
ax.set_title('3D Scatter Plot')
plt.show()
```

## 3、3D等高线图

ax.contour3D()函数可以用来创建三维等高线图，该函数要求输入数据均采用二维网格格式的矩阵坐标。同时，它可以在每一个网格点(x,y)处计算一个z值。

下面示例展示了如何绘制三维正弦等高线图，代码如下：

```python
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt

def f(x,y):
    return np.sin(np.sqrt(x**2 + y**2))

# 构建x,y数据
x = np.linspace(-6,6,30)
y = np.linspace(-6,6,30)

# 将数据网格化处理
X,Y = np.meshgrid(x,y)
Z = f(X,Y)

# 创建3D绘图区域
fig = plt.figure()
ax = plt.axes(projection='3d')
# 50表示：在z轴方向上，等高线的高度层级，binary表示：颜色从白色变成黑色
ax.contour3D(X,Y,Z,50,cmap='cividis')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('3D contour')
plt.show()

```

## 4、3D线框图

线框图同样要采用二维网格的形式数据，与绘制等高线类似。

线框图可以将数据投影在指定的三维表面上，并输出可视化程度较高的三维效果图。通过plot_wireframe()方法实现3D线框图，代码如下：

```python
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt

# 定义函数
def f(x,y):
    return np.sin(np.sqrt(x**2 + y**2))

# x,y数据
x = np.linspace(-6,6,30)
y = np.linspace(-6,6,30)
# 网格化数据
X,Y =np.meshgrid(x,y)
Z = f(X,Y)
print(X.shape, Y.shape, Z.shape)
print(X,Y,Z)
# 绘制图像
fig = plt.figure()
ax = plt.axes(projection='3d')
# 线框图 plot_wireframe()
ax.plot_wireframe(X,Y,Z,color='black')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('3D wireframe')
plt.show()
```

## 5、3D 曲面图

曲面图表示一个指定的因变量y和两个自变量的函数关系。

3D曲面图是一个三维图像，非常类似线框图，不同之处在于，线框图的每个面用多边形填充而成。plot_surface()函数用来绘制3D曲面图，该函数需要接收3个参数x,y,z。示例代码如下：

```python
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt


# 求向量外积
x = np.outer(np.linspace(-2,2,30),np.ones(30))
# 矩阵转置
y = x.copy().T
# 计算z
z = np.cos(x**2 + y**2)

# 绘制曲面图
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_surface(x,y,z,cmap='RdYlBu_r',edgecolor='none')
ax.set_title('3D Surface plot')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
plt.show()
```
